```python
#@title Imports
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

##Clean of data


```python
folder_with_data='data'
name_file='raw_data.csv'
input_file=os.path.join(folder_with_data,name_file)
df=pd.read_csv(input_file)
print('shape',df.shape)
print('initial colums \n',df.columns)
#select the columns we don't want
columns_to_drop=[
                 'Título', #duplicated and the other has better naming
                 'Contrato e Incidencias',
                 'url_with_details',
                 'Nº de expediente:',#duplicated
                 'Duración:',#this column is 'Fecha fin actual' - 'Fecha de inicio' 
                 'Descarga de pliegos:',
                 'Incidencias del contrato:',
                 'Importe bruto adjudicación(€)',
                 'Importe neto adjudicación(€)',
                 'Valor estimado contrato neto(€)',
                 'Importe neto licitación:',
                 'Importe neto adjudicación:',
                 'Valor estimado contrato neto:',
                 'Unnamed: 26' #only temporal
                 ]
#columns in datatime format
datetime_columns=['Fecha de Contrato','Fecha de inicio', 'Fecha fin actual']
#numeric columns
numeric_columns=[ 
                 'Importe bruto licitación:',
                 'Importe bruto adjudicación:',
                 ]
```

```python
#drop innecesary columns
df.drop(columns_to_drop,axis=1,inplace=True)
df.drop_duplicates(inplace=True)#sometimes the last row is the first in the next page
```

```python
df.head()
```

```python
df.info()
```

```python
#set the correct dtype per columns
for datetime_column in datetime_columns:
    if 'Fecha de Contrato'==datetime_column:
        format='%d.%m.%Y'
    else:
        format='%d/%m/%Y'
    df[datetime_column]=pd.to_datetime(df[datetime_column],format=format,errors='coerce')
for numeric_column in numeric_columns:
    df[numeric_column]=df[numeric_column].str.replace("€","")
    df[numeric_column]=df[numeric_column].str.replace(".","")
    df[numeric_column]=df[numeric_column].str.replace(",",".")
    df[numeric_column]=df[numeric_column].astype(float)
```

###creating new features

```python
df['duration_days']=(df['Fecha fin actual']-df['Fecha de inicio']).dt.days
df['Importe bruto adjudicación per day']=df['Importe bruto adjudicación:']/df['duration_days']
df['% bruto respect licitacion']=df['Importe bruto adjudicación:']/df['Importe bruto licitación:']
```

##Checking outliers

```python
#@title Outliers
for x in df.columns:
    if pd.api.types.is_numeric_dtype(df.loc[:,x]):
        q75,q25 = np.percentile(df.loc[:,x],[75,25])
        intr_qr = q75-q25

        max = q75+(1.5*intr_qr)
        min = q25-(1.5*intr_qr)

        df.loc[df[x] < min,x] = np.nan
        df.loc[df[x] > max,x] = np.nan
df.isnull().sum().sort_values()
```

We have a high number of outlier in 'Importe bruto adjudicación:','Importe bruto licitación:' both feature have a high correlation

###study of outliers

```python
def showing_n_outliers_per_variable_x(df,column,n):
    return df[['Importe bruto adjudicación:','Importe bruto licitación:']].isnull().groupby(df[column]).sum().astype(int).reset_index().sort_values('Importe bruto adjudicación:',ascending=False).head(n)
def box_plot_per_variable_x_importe_adjudicación(df,column):
    return px.box(df, x=column,y="Importe bruto adjudicación:")
    
```

```python
interest_columns_to_study_of_outliers=['Destino','Número de licitadores:']
```

```python
for column in interest_columns_to_study_of_outliers:
    df_aux=showing_n_outliers_per_variable_x(df,column,10)
    print(column,'\n',df_aux)
    fig=box_plot_per_variable_x_importe_adjudicación(df,column)
    fig.show()
```
